### Data exploration

```python
train_df=pd.read_csv('train.csv')
train_df.head(3)
```

```python
val_df=pd.read_csv('validation.csv')
val_df.head(3)
```

```python
train_df=train_df.drop('Units_sold',axis=1)
train_df.head()
```

```python
val_df=val_df.drop('Units_sold',axis=1)
val_df.head()
```

```python
Rating_ratio=[]
for row in train_df.values:
    if(row[3]+row[4]==0):
        if(row[6]+row[7]==0):
             Rating_ratio.append(0.0)
        else:
            Rating_ratio.append(-99999)
    else:
        Rating_ratio.append((int(row[6]+row[7]))/(int(row[3]+row[4])))
        
        
        
max_rating=max(Rating_ratio)
for x in range(len(Rating_ratio)):
    if(Rating_ratio[x]==-99999):
        Rating_ratio[x]=max_rating
        

train_df['good_by_bad_Rating']=Rating_ratio 

train_df=train_df.drop(columns=['1_Star_Rating','2_Star_Rating','3_Star_Rating','4_Star_Rating','5_Star_Rating'],axis=1)
       
train_df.head().transpose()    
```

```python
Rating_ratio=[]
for row in val_df.values:
    if (row[3]+row[4]==0):
        if (row[6]+row[7]==0):
            Rating_ratio.append(0.0)
        else:
            Rating_ratio.append(-99999)
    else:
        Rating_ratio.append((int(row[6]+row[7]))/(int(row[4]+row[3])));

max_rating=max(Rating_ratio)
for x in range(len(Rating_ratio)):
    if (Rating_ratio[x]==-99999):
        Rating_ratio[x]=max_rating

val_df['good_by_bad_Rating']=Rating_ratio
                            
val_df=val_df.drop(columns=['1_Star_Rating','2_Star_Rating','3_Star_Rating','4_Star_Rating','5_Star_Rating'],axis=1)
val_df.head().transpose()                             
```

### Encoding Cregorical variables

```python
#one hot encoding
train_df=pd.get_dummies(train_df,columns=['Segment'],prefix='Segment',drop_first=True)
val_df=pd.get_dummies(val_df,columns=['Segment'],prefix='Segment',drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train=train_df.drop('Units_sold>1000',axis=1)
y_train=train_df['Units_sold>1000']
```

```python
X_val=val_df.drop('Units_sold>1000',axis=1)
y_val=val_df['Units_sold>1000']
```

### Training a Decision Tree Model

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model=GradientBoostingClassifier(random_state=42)
model.fit(X_train,y_train)

y_pred_train=model.predict(X_train)
y_pred_val=model.predict(X_val)
```

```python
from sklearn.metrics import f1_score


f1=f1_score(y_train,y_pred_train)
print(f'F1 score on train:{round(f1*100,2)}%')

f1=f1_score(y_val,y_pred_val)
print(f'F1 score on validation:{round(f1*100,2)}%')

```

```python
model=GradientBoostingClassifier(max_depth=9,min_samples_leaf=30,random_state=42)
model.fit(X_train,y_train)

y_pred_train=model.predict(X_train)
y_pred_val=model.predict(X_val)
```

```python
f1=f1_score(y_train,y_pred_train)
print(f'F1 score on train:{round(f1*100,2)}%')

f1=f1_score(y_val,y_pred_val)
print(f'F1 score on validation:{round(f1*100,2)}%')

```
